```python
import numpy as np
import gym
import lib
from importlib import reload
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
env=gym.make('MountainCar-v0')
```

```python
reload(lib)
res, model = lib.run(env, win=-110.0)
```

```python
state = env.reset()
sumr = 0
for i in range(3000):
    env.render()
    policy = model.getPolicy(state, 0.0)
    action = np.random.choice(3, p=policy)
    n_state, reward, done, info = env.step(action)    
    sumr += reward
    state = n_state    
    if (done):
        state = env.reset()
        print('Reward: %d' % (sumr))
        sumr = 0
```

```python
reload(lib)
model = lib.Model(env)
min_x, max_x = -1.2, 0.6
min_v, max_v = -0.07, 0.07
x = np.linspace(min_x, max_x, 40)
v = np.linspace(min_v, max_v, 60)
phi = []
k = 10
for vv in v:
    phi.append([])
    for xx in x:
        phi[-1].append(model.Featurizer.get_features([(xx, vv)])[k])        
        phi[-1][-1] *= model.Featurizer.std[k]
        phi[-1][-1] += model.Featurizer.mean[k]
phi = np.array(phi)
print(phi.shape)
d = phi
mx, mv = np.meshgrid(x, v)
CS = plt.contourf(mx, mv, d)
sx = model.Featurizer.samples[:, 0]
sy = model.Featurizer.samples[:, 1]
if (k >= 3):
    plt.scatter(sx[k - 3], sy[k - 3], c='k', marker='x')
plt.colorbar(CS)
plt.grid()
plt.show()
```
